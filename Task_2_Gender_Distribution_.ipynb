```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data_set.csv")
```

```python
# Display top 5 rows
df.head()
```

```python
gender_data = df['gender'].value_counts()
gender_data
```

* Gender Categories: The dataset includes two gender categories: 'Male' and 'Female'.
* Count Statistics: The counts for each gender are as follows:
* Male: 25
* Female: 15



```python
# Create a pie chart to visualize gender distribution
plt.figure(figsize=(6, 6))
plt.pie(gender_data, labels=gender_data.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'], startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Gender Distribution')
plt.show()
```

Proportion Representation: The pie chart shows the proportion of each gender in the dataset.
* Male: 62.5%
* Female: 37.5%

**Conclusion and Insights**
* Gender Distribution: The dataset has a higher proportion of males (62.5%) compared to females (37.5%).
* Visualization Choice: Both bar charts and pie charts are effective for visualizing gender distribution. The pie chart used here clearly shows the proportion of each gender.

**Insights :**
* More no.of males less no.of females
* Males are twice as much females
