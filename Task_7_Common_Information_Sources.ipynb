Objective : Analyze common sources participants
rely on for investment information.

Steps:
1. Explore Information Sources Column: Review
the column where participants indicated their
sources of investment information.

2. Identify Common Sources: Analyze the data to
identify and summarize the most common
sources participants rely on.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data_set.csv")
```

```python
# Display top 5 rows
df.head()
```

```python
# Display columns to locate the "Information Sources" column
df.columns
```

```python
# Analyze the "Source" column to find the most common investment information sources
common_sources = df['Source'].value_counts()

# Display the most common sources
common_sources
```

```python
# Data for the bar chart
sources = ['Newspapers and Magazines', 'Financial Consultants', 'Television', 'Internet', 'Social Media']
frequencies = [14, 12, 9, 8, 3]

# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(sources, frequencies, color=['blue', 'green', 'orange', 'red', 'purple'])
plt.title('Common Sources of Investment Information')
plt.xlabel('Sources')
plt.ylabel('Frequency')
plt.xticks(rotation=15)
plt.grid(axis='y')
plt.grid(False)

# Show the plot
plt.tight_layout()
plt.show()
```

**1. Newspapers and Magazines:**

Frequency: 14 mentions  
Insight : This traditional medium remains a trusted source for many participants, suggesting that they value in-depth analyses and expert opinions typically found in print.

**2. Financial Consultants:**

Frequency: 12 mentions   
Insight: The reliance on professional financial advice indicates a strong preference for personalized guidance when making investment decisions.

**3. Television:**

Frequency: 9 mentions  
Insight : Television continues to be a popular source, particularly for real-time updates and market news, appealing to those who prefer visual and auditory information.

**4. Internet:**

Frequency: 8 mentions  
Insight: The internet is increasingly becoming a go-to source, reflecting a shift towards digital platforms for quick access to information and diverse viewpoints.

**5. Social Media:**

Frequency: 3 mentions  
Insight: While less common, social media is emerging as a source, particularly among younger participants, indicating a growing trend towards informal and peer-driven information sharing.
